```python
import numpy as np
import os
import pandas as pd
import csv
from matplotlib import pyplot as plt
import matplotlib.pyplot as pyplot
import PIL.Image as Image
from PIL import ImageOps
import glob
import math
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data", header = None )
```

```python
from sklearn import linear_model
df1 = df
df1 = df1.drop([0],axis=1)
y=df.iloc[:,-1]

df1 = df1.drop(df1.columns[-1],axis=1)

x=df1

reg = linear_model.LinearRegression()

reg.fit(x,y)
x_r_1 = reg.predict(x)
y_r = y - reg.predict(x)

plt.scatter(x_r_1 ,y_r ,color ='green')
plt.hlines(y = 0, xmin = min(x_r_1)-1, xmax = max(x_r_1)+1)
plt.title('Residual Plot against fitted values without attribute gender')
plt.xlabel('fitted values')
plt.ylabel('residuals')
plt.show()
```
